```python

def calc(a, b, c):
  import math as m

  if c == '+':
      return a + b
  elif c == '-':
      return a - b
  elif c == '*':
      return a * b
  elif c == '/':
      if b == 0:
          raise ZeroDivisionError("ZeroDivisionError : 0으로 나눌 수 없습니다.")
      return a / b
  elif c == '**':
      return m.pow(a,b)
  else:
      raise Exception("연산자는 +, -, *, /, ** 중 한가지만 입력하세요.")


while True:
    try:
        in1 = int(input("정수1을 입력하세요. : "))
        in2 = int(input("정수2를 입력하세요. : "))
        in3 = input("연산자를 입력하세요. : ")

        rst = calc(in1, in2, in3)

        print(f">>> {in1}{in3}{in2} 결과는 {rst} 입니다.")

    except ValueError:
        print('잘못된 입력입니다. 정수를 입력해주세요.')
    except Exception as ex:
        print(f'{ex}')

    q = input("계속 계산하시겠습니까?(Y/N)")
    if q.upper() == "N":
        break
    print("="*40)
```
